```python
import sklearn.datasets as skdatasets  # Concise import
import matplotlib.pyplot as plt

# Load the diabetes dataset
diabetes = skdatasets.load_diabetes()

# Extract features and target
X = diabetes.data
y = diabetes.target

# Feature names (more informative)
features = diabetes.feature_names
AGE = features.index("age")
MASS_INDEX = features.index("bmi")  # Using 'bmi' for Body Mass Index
BLOOD_PRESSURE = features.index("bp")

print("Data shape (samples, features):", X.shape)
print("Target shape (samples):", y.shape)

# Print the first 5 data points with all features
print("First 5 data points:\n", X[:5])

# Print the target values for the first 5 data points
print("Target values for first 5 data points:\n", y[:5])

# Get mass index of the second data point
second_mass_index = X[1, MASS_INDEX]
print("Mass index of the second data point:", second_mass_index)

# Get age for data points 1 and 2 (inclusive)
age_1_2 = X[1:3, AGE]
print("Age for data points 1 and 2:", age_1_2)

# Number of samples in the dataset
num_samples = diabetes.data.shape[0]

# Improved scatter plots with proper indexing
x = range(num_samples)  # Corrected x-axis for plots

plt.figure(figsize=(12, 4))  # Set figure size

plt.plot(x, X[:, MASS_INDEX], 'y-', label='Mass Index')
plt.plot(x, X[:, BLOOD_PRESSURE], 'b-', label='Blood Pressure')

plt.xlabel("Sample Index")  # Clearer axis labels
plt.ylabel("Feature Value")

plt.title("Mass Index and Blood Pressure Distribution")
plt.legend()
plt.grid(True)
plt.tight_layout()  # Adjust layout for readability
plt.show()

# Improved subplots with correct data access
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

for i, feature in enumerate([MASS_INDEX, AGE]):
    axs[i].scatter(X[:, feature], X[:, BLOOD_PRESSURE], c='g', marker='o', label=features[feature])
    axs[i].set_title(features[feature] + " vs. Blood Pressure")
    axs[i].set_xlabel(features[feature])
    axs[i].set_ylabel("Blood Pressure")
    axs[i].grid(True)

plt.legend()
plt.tight_layout()
plt.show()
```
